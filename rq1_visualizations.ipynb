```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
includes_drama = pd.read_csv('data/includes_drama.csv')
includes_comedy = pd.read_csv('data/includes_comedy.csv')
includes_thriller = pd.read_csv('data/includes_thriller.csv')
includes_action = pd.read_csv('data/includes_action.csv')
includes_romance = pd.read_csv('data/includes_romance.csv')
includes_documentation = pd.read_csv('data/includes_documentation.csv')
includes_crime = pd.read_csv('data/includes_crime.csv')
```

```python
sns.relplot(data = includes_drama, x = "prop_female", y = "tmdb_popularity", kind = "scatter")
plt.xlabel("Proportion of Female Crew/Cast")
plt.ylabel("Popularity Score")
sns.relplot(data = includes_comedy, x = "prop_female", y = "tmdb_popularity", kind = "scatter")
plt.xlabel("Proportion of Female Crew/Cast")
plt.ylabel("Popularity Score")
sns.relplot(data = includes_thriller, x = "prop_female", y = "tmdb_popularity", kind = "scatter")
plt.xlabel("Proportion of Female Crew/Cast")
plt.ylabel("Popularity Score")
sns.relplot(data = includes_action, x = "prop_female", y = "tmdb_popularity", kind = "scatter")
plt.xlabel("Proportion of Female Crew/Cast")
plt.ylabel("Popularity Score")
sns.relplot(data = includes_romance, x = "prop_female", y = "tmdb_popularity", kind = "scatter")
plt.xlabel("Proportion of Female Crew/Cast")
plt.ylabel("Popularity Score")
sns.relplot(data = includes_documentation, x = "prop_female", y = "tmdb_popularity", kind = "scatter")
plt.xlabel("Proportion of Female Crew/Cast")
plt.ylabel("Popularity Score")
sns.relplot(data = includes_crime, x = "prop_female", y = "tmdb_popularity", kind = "scatter")
plt.xlabel("Proportion of Female Crew/Cast")
plt.ylabel("Popularity Score")
```
